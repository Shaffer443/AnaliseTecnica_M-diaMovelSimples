```python
#Instalações e Importações:

!pip install yfinance
import yfinance as yf
import pandas as pd
```

```python
#Criando duas variáveis para determinar o começo e o fim do periodo a ser analisado

inicio= "2022-01-01"
fim = "2022-09-10"

```

```python
#Iniciando a importação dos dados de ações. Como exemplo, so dados da Vale3.

vale3 = yf.download('VALE3.SA', start = inicio, end = fim)
```

```python
# Usando Pandas:

# Função head(), mostra as 5 priemiros valores:

vale3.head()
```

```python
# Usando Pandas:

# Função tail(), mostra os 5 últimos valores:

vale3.tail()
```

```python
#Criando as médias móveis.
#Média Móvel é uam ferramenta estatistica ultilizada para suavizar flutuação de dados temporais.
# Há varios tipos de médias, vamos implementar a média móveis simples. Que é a média de N valores passados.
# No exemplo móvel simpels de 20 periodos retorna a média dos ultimos X dias.

# Criando dois objetos: Um contendo a média móvel de 20 peridoos , e outros de 60 periodos.

sma20=vale3['Adj Close'].rolling(20).mean()
sma60=vale3['Adj Close'].rolling(60).mean()
```

```python
#Passando esses dois objetos acima para dentro do DataFrame de Vale 3:

vale3['MM20']= sma20
vale3['MM60'] = sma60
```

```python
# Inspecionar Dados:

vale3.tail()
```

```python
#Gráficos:

vale3[['Adj Close', 'MM20','MM60']].plot(figsize=(20,10));
```
